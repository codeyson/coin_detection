```python
from IPython.display import display, Image, clear_output, Javascript
import os
from ultralytics import YOLO
!yolo checks
```

```python
# Set root directory to your local path
ROOT_DIR = "data"
```

```python
# Load model
model_coin = YOLO("yolov8m.pt")
```

```python
# Train model
results = model_coin.train(
    data=os.path.join(ROOT_DIR, "data/data.yaml"),
    epochs=128
)
```

```python
# After training, load best weights
model = YOLO("runs/detect/train/weights/best.pt")
```

```python
# Run inference on test images
results = model.predict(
    source=os.path.join(ROOT_DIR, "test/images"),
    conf=0.25,
    save=True
)
```
